```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
nodes=36
scenarios=3
```

```python
trainset=pd.DataFrame()
c=1
arr=[]
for i in range(1,scenarios+1):
    for j in range(1,nodes+1):
        a=pd.read_csv("n"+str(j)+"_"+str(i)+".csv")
        tra=np.array(a.iloc[:nodes,1])
        arr=list(tra)
        arr.append(int(j))
        trainset[(c)]=arr
        c+=1
```
