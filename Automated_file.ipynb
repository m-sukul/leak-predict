```python
def extractp():
    #path="G:\\research\\auto\\fos_read\\"
    orig=pd.read_csv(path+'original.csv')
    diffs=pd.DataFrame()
    diffs['nodes']=range(1,nodes+1)
    for i in range(1,nodes+1):
        a=pd.read_csv(path+"n"+str(i)+"_1.csv")
        diff=np.array((orig.iloc[:nodes,1])-a.iloc[:nodes,1])
        diffs[str(i)]=diff
    diffs=diffs.set_index('nodes')
    
    sums=pd.DataFrame()
    sums['sum']=diffs.sum(axis=1)
    sums.sort_values('sum',inplace=True)
    sensors=np.array(sums.index)
    return sensors
```

```python
nodes=36
def get_sensors(limit=nodes):
    sensors=extractp()
    n=len(sensors)
    i=n//2 -1
    left=i-limit//2
    right=i+limit//2+1
    sensors=sensors[left:right]
    sensors=sensors-1
    return sensors.tolist()
    
    
```
